```python
fp_cookies_syncing = cookies_syncing[(cookies_syncing.share_fp_cookie) & (cookies_syncing.referred_by_third) 
                                     & (cookies_syncing.referrer.notna())].reset_index(drop = True)

# fp_cookies_syncing.groupby(['exp_type','exp_group', 'exp_index']).size().reset_index(name='count')


fp_cookies_syncing['host'] = list(map(lambda x: clean_host(x), fp_cookies_syncing.host.tolist()))
fp_cookies_syncing['referrer'] = list(map(lambda x: clean_host(x), fp_cookies_syncing.referrer.tolist()))

fp_cookies_syncing
temp = fp_cookies_syncing.groupby(['exp_group','host']).size().reset_index(name='count').sort_values(['exp_group','count'], ascending=False)
temp[temp.exp_group == 'c']
```
