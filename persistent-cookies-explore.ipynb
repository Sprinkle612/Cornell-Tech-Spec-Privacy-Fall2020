```python
import numpy as np
import pandas as pd
```

```python
cookies = pd.read_csv('distinct_cookies_update.csv')
```

```python
cookies.sort_values(by=['exp_type', 'exp_group', 'exp_index', 'value'], inplace=True)
cookies.shape
```

```python
v = cookies.value.value_counts()
```

```python
cookies_cleaned = cookies[cookies.value.isin(v.index[v.gt(1)])]
```

```python
cookies_cleaned = cookies_cleaned[(cookies_cleaned.value.str.len() > 5) & (~cookies_cleaned.value.isin(['.forbes.com', '.webmd.com', 'cookie', 'testWrite','NO_DATA', '%24NOT_FOUND']))]
```

```python
cookies_cleaned.shape
```

```python
helper = cookies_cleaned.groupby(['exp_group', 'exp_type', 'exp_index', 'value']).size().reset_index(name='counts')
helper = helper[helper.counts > 1]
final_table = cookies_cleaned[cookies_cleaned.value.isin(helper.value.tolist())]
```

```python
def extract_persistent_cookies(eg):
    tables = []
    for et in ['forbes', 'nyt', 'washington']:
        for i in range(1, 7):
            temp = final_table[(final_table.exp_group == eg) & (final_table.exp_index==i) & (final_table.exp_type == et)]
            helper = temp.groupby('value').size().reset_index(name='counts')
            helper = helper[helper.counts > 1]
            temp = temp[temp.value.isin(helper.value.tolist())]
            tables.append(temp)
    return pd.concat(tables)
```

```python
t = extract_persistent_cookies('t')
```
