```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
get_set_cookies = pd.read_csv('get_set_cookies.csv')
```

```python
get_set_cookies = get_set_cookies.drop(columns=['visit_id', 'name', 'value'])
```

```python
# third party scripts accessed cookies from different domains
diff_domains_subset = get_set_cookies[get_set_cookies.cleaned_host != get_set_cookies.script_host]
```

```python
diff_domains_subset[:5]
```

```python
counts_all = get_set_cookies.groupby(['type' ,'exp_group', 'origin_site', 'exp_index']).size().reset_index(name='counts')
```

```python
counts_all_t = counts_all[counts_all.exp_group == 't']
counts_all_c = counts_all[counts_all.exp_group == 'c']
```

```python
counts_all_t[counts_all_t.origin_site=='nytimes']
```

```python
counts_all_c[counts_all_c.origin_site=='nytimes']
```

```python
for site in counts_all.origin_site.unique():
    # GET
    get_t = counts_all_t[(counts_all_t.type == 'get') & (counts_all_t.origin_site==site)].counts.tolist()
    get_c = counts_all_c[(counts_all_c.type == 'get') & (counts_all_c.origin_site==site)].counts.tolist()
    print("Get "+site, stats.ttest_ind(get_t, get_c))
    # SET
    set_t = counts_all_t[(counts_all_t.type == 'set') & (counts_all_t.origin_site==site)].counts.tolist()
    set_c = counts_all_c[(counts_all_c.type == 'set') & (counts_all_c.origin_site==site)].counts.tolist()
    print("Set "+site, stats.ttest_ind(set_t, set_c))
        
```

```python
diff_domains_subset.groupby(['origin_site', 'type','exp_group'])['script_host', 'cookie_value'].nunique().reset_index()
```

```python
summary = diff_domains_subset.groupby(['exp_group', 'origin_site', 'script_host', 'type']).size().reset_index(name="count")
```

```python
summary_t = summary[summary.exp_group=='t']
summary_c = summary[summary.exp_group=='c']

compared_table = pd.merge(summary_c, summary_t, on=['origin_site', 'script_host', 'type'], how='outer', suffixes=('_control', '_treatment'))

compared_table.fillna(0, inplace=True)

compared_table.drop(columns=['exp_group_control', 'exp_group_treatment'], inplace=True)
```

```python
compared_table.to_csv('compare_set_get.csv', index=False)
```

```python
compared_table['diff'] = list(map(lambda a, b: abs(b-a), compared_table.count_control, compared_table.count_treatment))
```

```python
sort_compare = compared_table.sort_values(by=['origin_site', 'type', 'diff'], ascending=False)
```

```python
temp = sort_compare.groupby(['origin_site', 'script_host'])['count_control', 'count_treatment'].sum().reset_index()
```

```python
temp['abs_diff'] = list(map(lambda a, b: abs(b-a), temp.count_control, temp.count_treatment))
```

```python
temp.sort_values(by=['origin_site','abs_diff'], inplace=True, ascending=False)
```
