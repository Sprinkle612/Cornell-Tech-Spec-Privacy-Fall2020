```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
get_set_cookies = pd.read_csv('get_set_cookies.csv')
```

```python
get_set_cookies = get_set_cookies.drop(columns=['visit_id', 'name', 'value'])
```

```python
# third party scripts accessed cookies from different domains
diff_domains_subset = get_set_cookies[get_set_cookies.cleaned_host != get_set_cookies.script_host]
```

```python
counts_all = get_set_cookies.groupby(['type' ,'exp_group', 'origin_site', 'exp_index']).size().reset_index(name='counts')
```

```python
counts_all_t = counts_all[counts_all.exp_group == 't']
counts_all_c = counts_all[counts_all.exp_group == 'c']
```

```python
sites = counts_all.origin_site.unique()
```

```python
diff_domains_subset.groupby(['origin_site', 'type','exp_group'])['script_host', 'cookie_value'].nunique().reset_index()
```

```python
summary = diff_domains_subset.groupby(['exp_group', 'origin_site', 'script_host', 'type']).size().reset_index(name="count")
```

```python
summary_t = summary[summary.exp_group=='t']
summary_c = summary[summary.exp_group=='c']
```

```python
compared_table = pd.merge(summary_c, summary_t, on=['origin_site', 'script_host', 'type'], how='outer', suffixes=('_control', '_treatment'))
```

```python
compared_table.fillna(0, inplace=True)
```

```python
compared_table.drop(columns=['exp_group_control', 'exp_group_treatment'], inplace=True)
```

```python
compared_table.to_csv('compare_set_get.csv', index=False)
```
